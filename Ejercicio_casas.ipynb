```python
# dibujo de graficas histograma precio casas
plt.figure(figsize=(15, 7))

# Histograma
plt.subplot(1, 3, 1)
plt.hist(casas['preciocasa'], bins=50, edgecolor='black', alpha=0.7, color='skyblue')
plt.xlabel('Precio de la Casa', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Histograma del Precio de la Casa', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

```

```python
# Curva de densidad del precio
plt.figure(figsize=(10, 6))

casas['preciocasa'].plot(kind='kde', linewidth=2, color='darkblue')
plt.xlabel('Precio de la Casa', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.title('Curva de Densidad del Precio', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
from scipy import stats

plt.figure(figsize=(10, 6))

stats.probplot(casas['preciocasa'], dist="norm", plot=plt)
plt.title('Q-Q Plot - Verificación de Normalidad', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# test estadistico de normalidad
from scipy import stats

# Test de normalidad
estadistico, p_valor = stats.normaltest(casas['preciocasa'])

print("="*60)
print("TEST DE NORMALIDAD (D'Agostino y Pearson)")
print("="*60)
print(f"Estadístico: {estadistico:.4f}")
print(f"P-valor: {p_valor:.6f}")
print("\nInterpretación:")
if p_valor < 0.05:
    print("La distribución NO es normal (p < 0.05)")
    print("   Se rechaza la hipótesis de normalidad")
else:
    print("La distribución puede considerarse normal (p >= 0.05)")
print("="*60)
```

#Representar las relaciones entre el precio de la casa y el resto de características (numéricas) para ver si hay alguna relación fuerte o no. Diagrama de dispersión. Ver la matriz de correlación.

```python
# Identificar variables numéricas
numericas = casas.select_dtypes(include=['int64', 'float64']).columns.tolist()
vars_numericas = [col for col in numericas if col != 'preciocasa']

print("Variables numéricas (excluyendo precio):")
print(vars_numericas)
plt.show()
```

```python
# Diagrama de dispersión - Precio vs Área
plt.figure(figsize=(10, 6))

plt.scatter(casas['area'], casas['preciocasa'], alpha=0.5, s=30, color='blue')
plt.xlabel('Área', fontsize=12, fontweight='bold')
plt.ylabel('Precio Casa', fontsize=12, fontweight='bold')
plt.title('Precio vs Área', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Línea de tendencia
z = np.polyfit(casas['area'], casas['preciocasa'], 1)
p = np.poly1d(z)
plt.plot(casas['area'], p(casas['area']), "r--", alpha=0.8, linewidth=2, label='Tendencia')
plt.legend()

plt.tight_layout()
plt.show()

# Calcular correlación
correlacion = casas['area'].corr(casas['preciocasa'])
print(f"Correlación área-precio: {correlacion:.3f}")
```

```python
# Matriz de funcionamiento
# Calcular matriz de correlación
correlacion = casas[numericas].corr()

plt.figure(figsize=(14, 11))
sns.heatmap(correlacion, annot=True, fmt='.2f', cmap='coolwarm',
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
            center=0, vmin=-1, vmax=1)
plt.title('Matriz de Correlación entre Variables Numéricas',
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
#Análisis de correlaciones con el precio
print("="*60)
print("CORRELACIÓN CON EL PRECIO DE LA CASA")
print("="*60)

correlacion_precio = correlacion['preciocasa'].sort_values(ascending=False)
print("\nTodas las correlaciones con el precio:")
print(correlacion_precio)

print("\n" + "="*60)
print("INTERPRETACIÓN DE CORRELACIONES")
print("="*60)

# Clasificar por fuerza de correlación
print("\n Correlaciones FUERTES (|r| > 0.5):")
fuertes = correlacion_precio[(abs(correlacion_precio) > 0.5) & (correlacion_precio.index != 'preciocasa')]
if len(fuertes) > 0:
    for var, corr in fuertes.items():
        print(f"   {var}: {corr:.3f}")
else:
    print("  No hay correlaciones fuertes")

print("\n Correlaciones MODERADAS (0.3 < |r| ≤ 0.5):")
moderadas = correlacion_precio[(abs(correlacion_precio) > 0.3) & (abs(correlacion_precio) <= 0.5)]
if len(moderadas) > 0:
    for var, corr in moderadas.items():
        print(f"  • {var}: {corr:.3f}")
else:
    print("  No hay correlaciones moderadas")

print("\n Correlaciones DÉBILES (|r| ≤ 0.3):")
debiles = correlacion_precio[(abs(correlacion_precio) <= 0.3) & (correlacion_precio.index != 'preciocasa')]
if len(debiles) > 0:
    for var, corr in debiles.items():
        print(f"  ○ {var}: {corr:.3f}")
else:
    print("  No hay correlaciones débiles")

print("\n" + "="*60)
```

```python
#Función para detectar valores atípicos con IQR
def detectar_outliers_iqr(df, columna):
    """
    Detecta outliers usando el método del Rango Intercuartílico (IQR)

    Parámetros:
    - df: DataFrame
    - columna: nombre de la columna a analizar

    Retorna:
    - outliers: DataFrame con los outliers
    - limite_inferior: límite inferior (Q1 - 1.5*IQR)
    - limite_superior: límite superior (Q3 + 1.5*IQR)
    - Q1, Q3, IQR: estadísticos
    """
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]

    return outliers, limite_inferior, limite_superior, Q1, Q3, IQR

print("Función 'detectar_outliers_iqr' definida correctamente")
```

```python
#Detectar valores atípicos en variables continuas
print("="*60)
print("DETECCIÓN DE OUTLIERS (Método IQR)")
print("="*60)

# Variables continuas para detectar outliers
# Nota: Solo analizamos precio y área (las continuas)
vars_continuas = ['preciocasa', 'area']

resultados_outliers = {}

print("\nAnalizando variables continuas:")
print("-"*60)

for var in vars_continuas:
    outliers, lim_inf, lim_sup, Q1, Q3, IQR = detectar_outliers_iqr(casas_encoded, var)

    resultados_outliers[var] = {
        'outliers': outliers,
        'limite_inferior': lim_inf,
        'limite_superior': lim_sup,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'n_outliers': len(outliers)
    }

    print(f"\n {var.upper()}:")
    print(f"   Q1 (percentil 25):    {Q1:>12,.2f}")
    print(f"   Q3 (percentil 75):    {Q3:>12,.2f}")
    print(f"   IQR (Q3 - Q1):        {IQR:>12,.2f}")
    print(f"   Límite inferior:      {lim_inf:>12,.2f}")
    print(f"   Límite superior:      {lim_sup:>12,.2f}")
    print(f"   Outliers detectados:  {len(outliers):>12d}")
    print(f"   Porcentaje:           {(len(outliers)/len(casas_encoded))*100:>11.2f}%")

print("\n" + "="*60)
```

```python
#Eliminar valores atípicos
print("="*60)
print("ELIMINACIÓN DE OUTLIERS")
print("="*60)

print(f"\nFilas ANTES de eliminar outliers: {casas_encoded.shape[0]}")

# Crear copia del dataset
casas_sin_outliers = casas_encoded.copy()

# Eliminar outliers de cada variable continua
for var in vars_continuas:
    lim_inf = resultados_outliers[var]['limite_inferior']
    lim_sup = resultados_outliers[var]['limite_superior']

    # Filtrar datos dentro de los límites
    casas_sin_outliers = casas_sin_outliers[
        (casas_sin_outliers[var] >= lim_inf) &
        (casas_sin_outliers[var] <= lim_sup)
    ]

    print(f"\nDespués de eliminar outliers de '{var}':")
    print(f"   Filas restantes: {casas_sin_outliers.shape[0]}")

filas_eliminadas = casas_encoded.shape[0] - casas_sin_outliers.shape[0]

print("\n" + "-"*60)
print(f"Filas DESPUÉS de eliminar outliers: {casas_sin_outliers.shape[0]}")
print(f"Filas eliminadas: {filas_eliminadas}")
print(f"Porcentaje eliminado: {(filas_eliminadas/casas_encoded.shape[0])*100:.2f}%")

print("\nOutliers eliminados correctamente")
print("Dataset limpio: 'casas_sin_outliers'")
print("="*60)
```
