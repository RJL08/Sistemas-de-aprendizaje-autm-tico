```python
#AÃ±adir una nueva columna booleana para ver si el pasajero era menor de
#edad o no.
df["Menor"] = df["Age"]<18
df.head()
```

```python
# Borrar la columna y volver a crearla, pero esta vez que sea con valores Si o No
df.drop(columns="Menor", inplace=True)
df["EsMenor"] = ["Si" if edad < 18 else "No" for edad in df["Age"]]
df
```

```python
#porcentaje de menores de edad que sobreviveron en cada clase, en primer lugar, hay que hacerlo sin bucle y despues con bucle
df[df["EsMenor"]== "Si"].groupby("Pclass")["Survived"].mean()
```

```python
#Con bucle del anterior
menores = df[df["EsMenor"] == "Si"]
porcentajes = {}

for clase in sorted(menores["Pclass"].unique()):
  grupo = menores[menores["Pclass"] == clase]
  porcentaje = grupo["Survived"].mean() * 100
  porcentajes[clase] = round(porcentaje, 2)

print(porcentajes)

```
