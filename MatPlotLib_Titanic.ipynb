```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('sample_data/titanic.csv')
df
```

```python
#Crea un histograma con las edades de los pasajeros. Elimina los valores nulos
#y pon el histograma de color skyblue y con un borde negro. El resultado debe
#ser algo parecido a esto
plt.hist(df['Age'].dropna(), color='skyblue', edgecolor='black', bins = 20)
plt.xlabel('Edad')
plt.ylabel('Numeros de pasajeros')
plt.title('Distribución de edad de los pasajeros')
plt.show()
```

```python
# Crea un diagrama de barras horizontales con el número de personas de cada clase.
plt.barh(df['Pclass'].value_counts().sort_index().index.astype(str),
         df['Pclass'].value_counts().sort_index().values, color = "pink" , edgecolor= "black")
plt.xlabel('Numeros de pasajeros')
plt.ylabel('Clase')
plt.title('Número de pasajeros por clase')
plt.show()
```

```python
# crea un diagrama con el numero de personas de cada clase pero con un diagrama de sectores
plt.pie(df['Pclass'].value_counts().sort_index().values,
        labels = df['Pclass'].value_counts().sort_index().index.astype(str), autopct='%1.1f%%')
plt.title('Número de pasajeros por clase')
plt.show()
```
