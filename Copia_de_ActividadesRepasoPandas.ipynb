```python
import pandas as pd
```

```python
#importar el fichero
datos = pd.read_csv('sample_data/grades.csv')
datos
```

```python
datos.info()
```

```python

datos.index = ["Alan Turing", "Claude Shannon", "John McCarthy", "Marvin Minsky"]
datos.insert(0, "DNI", ["11222333A", "22333222B", "33222333C", "44333222D"])
datos.insert(1, "EDAD", [23, 23, 36, 36])
datos = datos[["DNI", "EDAD", "PIA", "SAA", "MIA", "SBD", "BDA"]]
datos
```

```python
datos['Otro'] = [1,1,1,1]
datos
```

```python
PIA_min = datos['PIA'].min()
PIA_max = datos['PIA'].max()


datos["PIA_nom"]= (datos["PIA"]-PIA_min)/(PIA_max-PIA_min)
datos
```

```python
#borramos la columna otro
datos = datos.drop(columns=['Otro'])
datos
```

```python
datos = datos.drop(columns=['EDAD','PIA_nom'])
datos
```
